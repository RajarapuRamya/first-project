```python
def group_anagrams(words):
    groups = {}

    for word in words:
        count = [0] * 26

        for char in word:
            index = ord(char) - ord('a')
            count[index] += 1

        key = tuple(count)

        if key in groups:
            groups[key].append(word)
        else:
            groups[key] = [word]

    return list(groups.values())


words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(group_anagrams(words))
```

```python
def group_anagrams(words):
    groups = {}

    for word in words:
        # Create sorted key
        key = "".join(sorted(word))

        # Add word to corresponding group
        if key in groups:
            groups[key].append(word)
        else:
            groups[key] = [word]

    # Convert dictionary values into a list
    result = []

    for group in groups.values():
        result.append(group)

    return result


# Driver Code
words = ["eat", "tea", "tan", "ate", "nat", "bat"]

result = group_anagrams(words)

print(result)
```
