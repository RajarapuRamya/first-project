```python
# ============================================
# Employee Management System
# ============================================

# Taking employee details dynamically using input()

employee_name = input("Enter Employee Name: ")          # String
employee_id = int(input("Enter Employee ID: "))         # Integer
company_name = input("Enter Company Name: ")            # String
department = input("Enter Department: ")                # String
designation = input("Enter Designation: ")              # String
monthly_salary = float(input("Enter Monthly Salary: ")) # Float
years_experience = float(input("Enter Years of Experience: ")) # Float
work_location = input("Enter Work Location: ")          # String

# Boolean value
employee_active = input("Is Employee Active? (True/False): ")

# Convert string input to Boolean
employee_active = employee_active.lower() == "true"

# Calculating yearly salary
yearly_salary = monthly_salary * 12

# ============================================
# Display Employee Details
# ============================================

print("\n")
print("==========================================")
print("        EMPLOYEE DETAILS CARD")
print("==========================================")

print(f"Employee Name       : {employee_name}")
print(f"Employee ID         : {employee_id}")
print(f"Company Name        : {company_name}")
print(f"Department          : {department}")
print(f"Designation         : {designation}")
print(f"Monthly Salary      : ₹{monthly_salary}")
print(f"Yearly Salary       : ₹{yearly_salary}")
print(f"Years of Experience : {years_experience}")
print(f"Work Location       : {work_location}")
print(f"Employee Active     : {employee_active}")

print("==========================================")

# ============================================
# Display Data Types
# ============================================

print("\n")
print("==========================================")
print("           DATA TYPES")
print("==========================================")

print("employee_name       :", type(employee_name))
print("employee_id         :", type(employee_id))
print("company_name        :", type(company_name))
print("department          :", type(department))
print("designation         :", type(designation))
print("monthly_salary      :", type(monthly_salary))
print("yearly_salary       :", type(yearly_salary))
print("years_experience    :", type(years_experience))
print("work_location       :", type(work_location))
print("employee_active     :", type(employee_active))

print("==========================================")
```
