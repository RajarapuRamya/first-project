```python
# =========================================
# ATM Withdrawal Validation System
# =========================================

# Correct ATM PIN
CORRECT_PIN = "9381"

# Function to validate ATM transaction
def atm_withdrawal(balance, daily_limit,
                   withdrawn_today, entered_pin,
                   withdrawal_amount):

    # Check PIN
    if entered_pin == CORRECT_PIN:

        # Check withdrawal amount
        if withdrawal_amount > 0:

            # Check balance availability
            if withdrawal_amount <= balance:

                # Check daily withdrawal limit
                if (withdrawn_today + withdrawal_amount) <= daily_limit:

                    # Check minimum balance rule
                    remaining_balance = balance - withdrawal_amount

                    if remaining_balance >= 1000:
                        return (
                            "Transaction Successful",
                            remaining_balance
                        )

                    else:
                        return (
                            "Transaction Failed: Minimum balance rule violated",
                            balance
                        )

                else:
                    return (
                        "Transaction Failed: Daily limit exceeded",
                        balance
                    )

            else:
                return (
                    "Transaction Failed: Insufficient balance",
                    balance
                )

        else:
            return (
                "Transaction Failed: Invalid withdrawal amount",
                balance
            )

    else:
        return (
            "Transaction Failed: Invalid PIN",
            balance
        )


# =========================================
# User Inputs
# =========================================

name = input("Enter Account Holder Name: ")

balance = float(input("Enter Account Balance: "))

daily_limit = float(input("Enter Daily Withdrawal Limit: "))

withdrawn_today = float(input("Amount Withdrawn Today: "))

entered_pin = input("Enter ATM PIN: ")

withdrawal_amount = float(input("Enter Withdrawal Amount: "))


# =========================================
# Function Call
# =========================================

message, remaining_balance = atm_withdrawal(
    balance,
    daily_limit,
    withdrawn_today,
    entered_pin,
    withdrawal_amount
)


# =========================================
# Transaction Summary
# =========================================

print("\n================================")
print("ATM Transaction Summary")
print("================================")

print(f"Account Holder   : {name}")
print(f"Account Balance  : ₹{balance}")
print(f"Daily Limit      : ₹{daily_limit}")
print(f"Withdrawn Today  : ₹{withdrawn_today}")
print(f"Withdrawal Amount: ₹{withdrawal_amount}")

print("================================")
print(message)
print(f"Remaining Balance: ₹{remaining_balance}")
print("================================")
```
