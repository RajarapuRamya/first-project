```python
# Function to convert comma-separated string into list
def string_to_list(data):
    result = []
    word = ""

    for ch in data:
        if ch == ",":
            result.append(word)
            word = ""
        else:
            word += ch

    result.append(word)
    return result


# Function to convert rating strings into integer list
def ratings_to_int(rating_list):
    numbers = []

    for item in rating_list:
        num = 0
        for digit in item:
            num = num * 10 + (ord(digit) - ord('0'))
        numbers.append(num)

    return numbers


# Function to find highest rating employee
def highest_rating(names, ratings):
    highest = ratings[0]
    employee = names[0]

    for i in range(len(ratings)):
        if ratings[i] > highest:
            highest = ratings[i]
            employee = names[i]

    return employee, highest


# Function to find lowest rating employee
def lowest_rating(names, ratings):
    lowest = ratings[0]
    employee = names[0]

    for i in range(len(ratings)):
        if ratings[i] < lowest:
            lowest = ratings[i]
            employee = names[i]

    return employee, lowest


# Function to calculate average rating
def average_rating(ratings):
    total = 0

    for rating in ratings:
        total += rating

    avg = total / len(ratings)
    return avg


# Function to display employees above average
def above_average(names, ratings, avg):
    print("\nEmployees Above Average:")

    for i in range(len(ratings)):
        if ratings[i] > avg:
            print(names[i])


# Main Program
names_input = input("Employee Names: ")
ratings_input = input("Ratings: ")

names = string_to_list(names_input)

rating_strings = string_to_list(ratings_input)
ratings = ratings_to_int(rating_strings)

high_emp, high_rate = highest_rating(names, ratings)
low_emp, low_rate = lowest_rating(names, ratings)
avg = average_rating(ratings)

print("\nEmployee with Highest Rating :", high_emp, "(" + str(high_rate) + ")")
print("Employee with Lowest Rating :", low_emp, "(" + str(low_rate) + ")")
print("Average Rating :", avg)

above_average(names, ratings, avg)
```
