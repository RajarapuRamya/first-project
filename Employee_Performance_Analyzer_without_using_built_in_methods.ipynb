```python
# Convert names string into list
def get_names(name_string):
    return name_string.split(",")

# Convert ratings string into integer list
def get_ratings(rating_string):
    rating_list = rating_string.split(",")
    ratings = []

    for rating in rating_list:
        ratings.append(int(rating))

    return ratings

# Find highest rating employee
def highest_employee(names, ratings):
    highest = ratings[0]
    employee = names[0]

    for i in range(1, len(ratings)):
        if ratings[i] > highest:
            highest = ratings[i]
            employee = names[i]

    return employee, highest

# Find lowest rating employee
def lowest_employee(names, ratings):
    lowest = ratings[0]
    employee = names[0]

    for i in range(1, len(ratings)):
        if ratings[i] < lowest:
            lowest = ratings[i]
            employee = names[i]

    return employee, lowest

# Calculate average rating
def average_rating(ratings):
    total = 0

    for rating in ratings:
        total = total + rating

    average = total / len(ratings)
    return average

# Display employees above average
def above_average(names, ratings, average):
    print("\nEmployees Above Average:")

    for i in range(len(ratings)):
        if ratings[i] > average:
            print(names[i])

# Main Program
employee_names = input("Employee Names: ")
employee_ratings = input("Ratings: ")

names = get_names(employee_names)
ratings = get_ratings(employee_ratings)

high_name, high_rating = highest_employee(names, ratings)
low_name, low_rating = lowest_employee(names, ratings)

avg = average_rating(ratings)

print("\nEmployee with Highest Rating :", high_name, "(" + str(high_rating) + ")")
print("Employee with Lowest Rating :", low_name, "(" + str(low_rating) + ")")
print("Average Rating :", avg)

above_average(names, ratings, avg)
```
