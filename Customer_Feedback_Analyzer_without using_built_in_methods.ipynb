```python
# Function to count total words
def count_words(sentence):
    words = sentence.split()
    total = 0

    for word in words:
        total += 1

    return total


# Function to count total vowels
def count_vowels(sentence):
    vowels = "aeiouAEIOU"
    total = 0

    for ch in sentence:
        for vowel in vowels:
            if ch == vowel:
                total += 1
                break

    return total


# Function to find longest word
def longest_word(sentence):
    words = sentence.split()

    longest = words[0]

    for word in words:
        if len(word) > len(longest):
            longest = word

    return longest


# Function to reverse each word
def reverse_words(sentence):
    words = sentence.split()

    print("\nReversed Words :")

    for word in words:
        reverse = ""

        for i in range(len(word) - 1, -1, -1):
            reverse += word[i]

        print(reverse, end=" ")

    print()


# Function to display character frequencies
def character_frequency(sentence):
    checked = []

    print("\nCharacter Frequencies :")

    for ch in sentence:
        if ch != " " and ch not in checked:
            freq = 0

            for char in sentence:
                if ch.lower() == char.lower():
                    freq += 1

            print(ch.lower(), ":", freq)
            checked.append(ch)


# Main Program
feedback = input("Enter Feedback: ")

print("\nTotal Words :", count_words(feedback))
print("Total Vowels :", count_vowels(feedback))
print("Longest Word :", longest_word(feedback))

reverse_words(feedback)

character_frequency(feedback)
```
