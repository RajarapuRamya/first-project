```python
class FailException(Exception):
    pass

try:
    marks = int(input("Enter marks: "))

    if marks < 0 or marks > 100:
        raise ValueError

    if marks < 35:
        raise FailException

except ValueError:
    print("Invalid Marks")

except FailException:
    print("Student Failed")

else:
    print("Student Passed")

finally:
    print("Result Processing Completed")
```
