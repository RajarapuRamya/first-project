```python
# Smart Employee ID Generator using for loop

# Taking inputs
employee_name = input("Enter Employee Name: ")
joining_year = input("Enter Joining Year: ")
department_name = input("Enter Department Name: ")

# Variables
name_part = ""
year_part = ""
dept_part = ""

# Extract first 3 letters from employee name
count = 0
for ch in employee_name:
    if ch != " ":
        name_part += ch
        count += 1
    if count == 3:
        break

# Extract last 2 digits from joining year
for i in range(len(joining_year) - 2, len(joining_year)):
    year_part += joining_year[i]

# Extract first 3 letters from department name
count = 0
for ch in department_name:
    if ch != " ":
        dept_part += ch
        count += 1
    if count == 3:
        break

# Convert to uppercase
name_part = name_part.upper()
dept_part = dept_part.upper()

# Generate Employee ID
employee_id = name_part + year_part + dept_part

# Display result
print("Generated Employee ID :", employee_id)
```
