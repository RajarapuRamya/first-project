```python
# ================================
# Bank Loan Approval System
# ================================

# Function to check loan eligibility
def check_loan_approval(age, salary, credit_score, existing_loans, employment_status, loan_amount):

    # Age validation
    if age < 21 or age > 60:
        return "Rejected: Age criteria not satisfied"

    # Salary validation
    elif salary <= 40000:
        return "Rejected: Salary is too low"

    # Credit score validation
    elif credit_score < 750:
        return "Rejected: Poor credit score"

    # Existing loans validation
    elif existing_loans >= 2:
        return "Rejected: Too many existing loans"

    # Employment validation
    elif employment_status.lower() != "employed":
        return "Rejected: Applicant is unemployed"

    # Loan amount validation
    elif loan_amount > (salary * 10):
        return "Rejected: Loan amount exceeds eligibility"

    # If all conditions are satisfied
    else:
        return "Loan Approved"


# ================================
# User Inputs
# ================================

name = input("Enter Applicant Name: ")

age = int(input("Enter Age: "))

salary = float(input("Enter Monthly Salary: "))

credit_score = int(input("Enter Credit Score: "))

existing_loans = int(input("Enter Existing Loans Count: "))

employment_status = input("Enter Employment Status (employed/unemployed): ")

loan_amount = float(input("Enter Loan Amount Required: "))


# ================================
# Function Call
# ================================

result = check_loan_approval(
    age,
    salary,
    credit_score,
    existing_loans,
    employment_status,
    loan_amount
)


# ================================
# Final Output
# ================================

print("\n==============================")
print("Applicant Summary")
print("==============================")

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Monthly Salary: ₹{salary}")
print(f"Credit Score: {credit_score}")
print(f"Existing Loans: {existing_loans}")
print(f"Employment Status: {employment_status}")
print(f"Loan Amount Required: ₹{loan_amount}")

print("\n==============================")
print(result)
print("==============================")
```
