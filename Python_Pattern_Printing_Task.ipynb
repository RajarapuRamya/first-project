```python
#Right-Aligned Increasing Triangle
rows = 5

for outer_loop in range(1, rows + 1):

    # Inner loop for spaces
    for inner_loop in range(rows - outer_loop):
        print(" ", end="")

    # Inner loop for stars
    for inner_loop in range(outer_loop):
        print("*", end="")

    print()
```

```python
#Right-Aligned Decreasing Triangle
rows = 5

for outer_loop in range(rows):

    # Inner loop for spaces
    for inner_loop in range(outer_loop):
        print(" ", end="")

    # Inner loop for stars
    for inner_loop in range(rows - outer_loop):
        print("*", end="")

    print()
```

```python
#hollow square pattern
rows = 4
cols = 4

for outer_loop in range(rows):
    for inner_loop in range(cols):

        if outer_loop == 0 or outer_loop == rows - 1:
            print("*", end="")
        elif inner_loop == 0 or inner_loop == cols - 1:
            print("*", end="")
        else:
            print(" ", end="")

    print()
```

```python
#center-aligned pyramid
rows = 4

for outer_loop in range(1, rows + 1):

    # Print spaces
    for inner_loop in range(rows - outer_loop):
        print(" ", end="")

    # Print stars
    for inner_loop in range(2 * outer_loop - 1):
        print("*", end="")

    print()
```

```python
#inverted center-aligned
rows = 4

for outer_loop in range(rows):

    # Print spaces
    for inner_loop in range(outer_loop):
        print(" ", end="")

    # Print stars
    for inner_loop in range(2 * (rows - outer_loop) - 1):
        print("*", end="")

    print()
```

```python
#Increasing Triangle
rows = 5

for outer_loop in range(1, rows + 1):
    for inner_loop in range(outer_loop):
        print("*", end="")
    print()
```

```python
#Decreasing Triangle
rows = 5

for outer_loop in range(rows, 0, -1):
    for inner_loop in range(outer_loop):
        print("*", end="")
    print()
```

```python
#Diamond
rows = 5

# Upper Part
for outer_loop in range(1, rows + 1):

    for inner_loop in range(rows - outer_loop):
        print(" ", end="")

    for inner_loop in range(2 * outer_loop - 1):
        print("*", end="")

    print()

# Lower Part
for outer_loop in range(rows - 1, 0, -1):

    for inner_loop in range(rows - outer_loop):
        print(" ", end="")

    for inner_loop in range(2 * outer_loop - 1):
        print("*", end="")

    print()
```

```python
#Number Triangle
rows = 5

for outer_loop in range(1, rows + 1):

    for inner_loop in range(1, outer_loop + 1):
        print(inner_loop, end="")

    print()
```

```python
#Floyd's Triangle
rows = 5
number = 1

for outer_loop in range(1, rows + 1):

    for inner_loop in range(outer_loop):
        print(number, end=" ")
        number += 1

    print()
```

```python
#Pascal's Triangle
rows = 5

for outer_loop in range(rows):

    number = 1

    for inner_loop in range(rows - outer_loop):
        print(" ", end="")

    for inner_loop in range(outer_loop + 1):
        print(number, end=" ")
        number = number * (outer_loop - inner_loop) // (inner_loop + 1)

    print()
```
