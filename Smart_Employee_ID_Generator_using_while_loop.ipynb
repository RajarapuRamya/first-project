```python
# Smart Employee ID Generator using while loop

employee_name = input("Enter Employee Name: ")
joining_year = input("Enter Joining Year: ")
department_name = input("Enter Department Name: ")

# Extract first 3 letters using while loop
name_part = ""
i = 0

while i < 3 and i < len(employee_name):
    if employee_name[i] != " ":
        name_part += employee_name[i]
    i += 1

dept_part = ""
j = 0

while j < 3 and j < len(department_name):
    if department_name[j] != " ":
        dept_part += department_name[j]
    j += 1

# Last 2 digits of year
year_part = joining_year[-2:]

# Convert to uppercase
employee_id = (name_part + year_part + dept_part).upper()

print("Generated Employee ID :", employee_id)
```
