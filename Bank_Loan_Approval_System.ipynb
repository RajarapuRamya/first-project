```python
# Bank Loan Approval System

print("===== Bank Loan Approval System =====")

# User Inputs
name = input("Enter Applicant Name: ")

age = int(input("Enter Age: "))
salary = float(input("Enter Monthly Salary: "))
credit_score = int(input("Enter Credit Score: "))
existing_loans = int(input("Enter Existing Loans Count: "))
employment_status = input("Enter Employment Status (employed/unemployed): ").lower()
loan_amount = float(input("Enter Loan Amount Required: "))

print("\n===== Loan Verification Result =====")

# Loan Approval Logic
if age >= 21 and age <= 60:

    if salary > 40000:

        if credit_score >= 750:

            if existing_loans < 2:

                if employment_status == "employed":

                    # Loan eligibility check
                    max_loan = salary * 10

                    if loan_amount <= max_loan:
                        print("Loan Approved")
                    else:
                        print("Rejected: Loan amount exceeds eligibility")

                else:
                    print("Rejected: Applicant is unemployed")

            else:
                print("Rejected: Too many existing loans")

        else:
            print("Rejected: Poor credit score")

    else:
        print("Rejected: Salary is too low")

else:
    print("Rejected: Age criteria not satisfied")

# Applicant Summary
print("\n===== Applicant Summary =====")
print("Applicant Name :", name)
print("Age            :", age)
print("Monthly Salary :", salary)
print("Credit Score   :", credit_score)
print("Existing Loans :", existing_loans)
print("Employment     :", employment_status)
print("Loan Amount    :", loan_amount)
```
