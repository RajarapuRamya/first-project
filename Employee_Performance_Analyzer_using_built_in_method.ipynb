```python
# Function to convert names string into list
def get_employee_names(names_string):
    return names_string.split(",")

# Function to convert ratings string into list of integers
def get_employee_ratings(ratings_string):
    ratings = ratings_string.split(",")
    return [int(rating) for rating in ratings]

# Function to find employee with highest rating
def highest_rating_employee(names, ratings):
    max_rating = max(ratings)
    index = ratings.index(max_rating)
    return names[index], max_rating

# Function to find employee with lowest rating
def lowest_rating_employee(names, ratings):
    min_rating = min(ratings)
    index = ratings.index(min_rating)
    return names[index], min_rating

# Function to calculate average rating
def average_rating(ratings):
    return sum(ratings) / len(ratings)

# Function to display employees above average
def employees_above_average(names, ratings, avg):
    print("\nEmployees Above Average:")
    for i in range(len(ratings)):
        if ratings[i] > avg:
            print(names[i])

# Main Program
names_input = input("Employee Names: ")
ratings_input = input("Ratings: ")

names = get_employee_names(names_input)
ratings = get_employee_ratings(ratings_input)

highest_name, highest_rating = highest_rating_employee(names, ratings)
lowest_name, lowest_rating = lowest_rating_employee(names, ratings)
avg = average_rating(ratings)

print(f"\nEmployee with Highest Rating : {highest_name} ({highest_rating})")
print(f"Employee with Lowest Rating : {lowest_name} ({lowest_rating})")
print(f"Average Rating : {avg}")

employees_above_average(names, ratings, avg)
```
