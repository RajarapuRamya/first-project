```python
# Convert names string into list
def get_names(names_string):
    return names_string.split(",")

# Convert ratings string into integer list
def get_ratings(ratings_string):
    ratings = ratings_string.split(",")

    for i in range(len(ratings)):
        ratings[i] = int(ratings[i])

    return ratings

# Find highest rating employee
def highest_employee(names, ratings):
    highest = max(ratings)
    index = ratings.index(highest)
    return names[index], highest

# Find lowest rating employee
def lowest_employee(names, ratings):
    lowest = min(ratings)
    index = ratings.index(lowest)
    return names[index], lowest

# Calculate average rating
def average_rating(ratings):
    return sum(ratings) / len(ratings)

# Display employees above average
def employees_above_average(names, ratings, avg):
    print("\nEmployees Above Average:")

    for i in range(len(names)):
        if ratings[i] > avg:
            print(names[i])

# Main Program
employee_names = input("Employee Names: ")
employee_ratings = input("Ratings: ")

names = get_names(employee_names)
ratings = get_ratings(employee_ratings)

highest_name, highest_rating = highest_employee(names, ratings)
lowest_name, lowest_rating = lowest_employee(names, ratings)

avg = average_rating(ratings)

print("\nEmployee with Highest Rating :", highest_name, f"({highest_rating})")
print("Employee with Lowest Rating :", lowest_name, f"({lowest_rating})")
print("Average Rating :", avg)

employees_above_average(names, ratings, avg)
```
