```python
# Smart Password Validator

password = input("Enter Password: ")

print("\nChecking Password :", password)

# Variables
upper = 0
lower = 0
digit = 0
special = 0

# Using FOR loop to check password contents
for ch in password:

    if ch.isupper():
        upper += 1

    elif ch.islower():
        lower += 1

    elif ch.isdigit():
        digit += 1

    else:
        special += 1


# Using WHILE loop to check minimum length
length = 0
i = 0

while i < len(password):
    length += 1
    i += 1


# String Slicing
print("\nFirst 3 chars :", password[:3])
print("Last 3 chars  :", password[-3:])


# Password Validation
if upper >= 1 and lower >= 1 and digit >= 1 and special >= 1 and length >= 8:
    print("\nStatus : Strong Password")
else:
    print("\nStatus : Weak Password")


# ---------------- BONUS CHALLENGES ----------------

# Mask middle characters
if len(password) > 2:
    masked = password[0] + "*" * (len(password)-2) + password[-1]
    print("\nMasked Password :", masked)

# Count total digits
print("Total Digits :", digit)

# Print uppercase letters
print("Uppercase Letters :", end=" ")
for ch in password:
    if ch.isupper():
        print(ch, end=" ")

# Reverse first half using slicing
half = len(password) // 2
reverse_half = password[:half][::-1]
print("\nReversed First Half :", reverse_half)

# Starts with uppercase and ends with digit
if password[0].isupper() and password[-1].isdigit():
    print("Starts with uppercase and ends with digit")
else:
    print("Does not satisfy start/end condition")
```
