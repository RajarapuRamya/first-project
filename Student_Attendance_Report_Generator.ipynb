```python
with open("attendance.txt", "r") as file:
    records = file.readlines()

total_students = 0
present_students = 0
absent_students = 0

for record in records:
    name, status = record.strip().split(",")

    total_students += 1

    if status == "Present":
        present_students += 1
    else:
        absent_students += 1

with open("report.txt", "w") as report:
    report.write("Attendance Report\n")
    report.write("-----------------\n")
    report.write(f"Total Students : {total_students}\n")
    report.write(f"Present Students : {present_students}\n")
    report.write(f"Absent Students : {absent_students}\n")

print("Report generated!")
```
